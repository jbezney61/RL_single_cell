```python
import pickle
```

```python
# dqn_search_results_CVCL_0292_..._debug.pkl is generated by running the RL with the parameters you see in rl_utils and rl_training, 10K iterations
# dqn_search_results_CVCL_0292_..._mainrun.pkl is generated by running the RL with the parameters you see in rl_utils and rl_training, 2M iterations
# beam searches here have no treshold (like it was put to -100)
# also tree search using dqn had no threshold (like it was put to -100) but then we filter the results to only keep those that reached the threshold
```

```python
with open('dqn_search_results_CVCL_0292_k1_n5_beam_debug.pkl', 'rb') as f:
    dqn_1_beam_debug = pickle.load(f)
with open('dqn_search_results_CVCL_0292_k190_n5_beam_debug.pkl', 'rb') as f:
    dqn_190_beam_debug = pickle.load(f)
with open('dqn_search_results_CVCL_0292_k1_n5_beam_mainrun.pkl', 'rb') as f:
    dqn_1_beam_mainrun = pickle.load(f)
with open('dqn_search_results_CVCL_0292_k190_n5_beam_mainrun.pkl', 'rb') as f:
    dqn_190_beam_mainrun = pickle.load(f)
with open('dqn_search_results_CVCL_0292_k4_n5_tree_mainrun.pkl', 'rb') as f:
    dqn_4_tree_mainrun = pickle.load(f)
with open('search_results_tree_k4_n5.pkl', 'rb') as f:
    classic_4_tree = pickle.load(f)
```

```python
dqn_4_tree_mainrun = dqn_4_tree_mainrun[(dqn_4_tree_mainrun['starting_distance']-dqn_4_tree_mainrun['final_distance'])/dqn_4_tree_mainrun['starting_distance'] > 0.5].copy()
classic_4_tree = classic_4_tree[classic_4_tree['covered_threshold'] == True].copy()
```

```python
print('using dqn_4_tree_mainrun')
print(((dqn_4_tree_mainrun['starting_distance'] - dqn_4_tree_mainrun['final_distance'])/dqn_4_tree_mainrun['starting_distance']).describe())
print('--'*50)
print('using classic_4_tree')
print(((classic_4_tree['starting_distance'] - classic_4_tree['final_distance'])/classic_4_tree['starting_distance']).describe())
print('--'*50)
print('using dqn_1_beam_debug')
print(((dqn_1_beam_debug['starting_distance'] - dqn_1_beam_debug['final_distance'])/dqn_1_beam_debug['starting_distance']).describe())
print('--'*50)
print('using dqn_190_beam_debug')
print(((dqn_190_beam_debug['starting_distance'] - dqn_190_beam_debug['final_distance'])/dqn_190_beam_debug['starting_distance']).describe())
print('--'*50)
print('using dqn_1_beam_mainrun')
print(((dqn_1_beam_mainrun['starting_distance'] - dqn_1_beam_mainrun['final_distance'])/dqn_1_beam_mainrun['starting_distance']).describe())
print('--'*50)
print('using dqn_190_beam_mainrun')
print(((dqn_190_beam_mainrun['starting_distance'] - dqn_190_beam_mainrun['final_distance'])/dqn_190_beam_mainrun['starting_distance']).describe())
print('--'*50)
```
