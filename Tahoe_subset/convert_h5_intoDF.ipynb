```python
import h5py
import pandas as pd
```

```python
# 1) Read your HDF5 PCA array into a DataFrame, re-using the same index as adata_sub.obs
with h5py.File("X_pca.h5", "r") as f:
    arr = f["X_pca"][:]   # loads into a NumPy array
```

```python
cols = [f"PC{i+1}" for i in range(arr.shape[1])]
df_pca = pd.DataFrame(arr, columns=cols)
```

```python
df_pca
```

```python
df_meta2 = pd.read_csv("sampled_metadata.csv")  
# ensure it has the same length
assert len(df_meta2) == len(df_pca)
```

```python
df_meta2
```

```python
df_pca["drugname_drugconc"] = df_meta2["drugname_drugconc"].values
df_pca["cell_line"] = df_meta2["cell_line"].values
```

```python
df_pca.to_csv('subsampled_PCA_metadata.csv')
```

```python
pc_cols = [c for c in df_pca.columns if c.startswith("PC")]

# 2) group by the two metadata columns and take the mean of each PC
df_grouped = (
    df_pca
    .groupby(["drugname_drugconc", "cell_line"])[pc_cols]
    .mean()
)
```

```python
df_grouped.to_csv('collapsed_subsampled_PCA_metadata.csv')
```
